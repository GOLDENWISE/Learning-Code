```julia
using Plots
plotlyjs()
 :circle
 

 f(x)=x^2 + 1
 g(x)=f(x)-1
 
 plot(f,-2,2, xlabel="x", ylabel="y", label="f(x)")
 plot!(g, label="g(x)")
```

```julia
using Plots

plotlyjs()

x = [2,3,4,1]
y = [8,3,1,2]
z = [4,5,3,4]

p1 = plot(x, y, seriestype=:scatter,
        title="Hello!", xlabel = "xlabel", ylabel="ylabel",
        xlims=(0.0, 6.5), ylims=(0.0,9.0))
p2 = plot(x, y, z, seriestype=:scatter,
        title="Hello!", ylabel="ylabel", zlabel="zlabel",
        xlims=(0.0, 12.5), ylims=(0.0,9.0), zlims=(0.0,5.0))
```

```julia
using Plots
plotlyjs

 f(x)= (x^2 - x - 6)/(x - 3) 
 g(x)= 5

plot(f, 0, 3, c=:blue, framestyle=:origin, label="f(x)")
plot!(g, 3, 5, c=:red, label="g(x)")
```
